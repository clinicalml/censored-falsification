```python
import pandas as pd
from src import LOG_DIR
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mmr_drop_file = "2023-05-16_00-42-57"
df_drop = pd.read_csv(os.path.join(LOG_DIR,"experiments","runs","mmr_drop", f"{mmr_drop_file}","type_1_indep_x.csv"))
config_drop = {"name": "Drop Censoring"}
drop_results = (df_drop, config_drop)


type1_indep_x_file = "2023-05-16_12-44-42"
df_type1_indep_x = pd.read_csv(os.path.join(LOG_DIR,"experiments","runs","type1_indep_x", f"{type1_indep_x_file}","type_1_indep_x.csv"))
config_type1_indep_x = {"name": "Censored MMR"}
type1_indep_results = (df_type1_indep_x, config_type1_indep_x)
```

```python
for res in [drop_results, type1_indep_results]:
    df, config = res
    df_ = df.groupby("obs_study_size")["reject"].mean()
    plt.plot(df_.index, df_.values, label=config["name"])
    plt.scatter(df_.index, df_.values, label=config["name"], marker="x")
    plt.xlabel("Observational Study Size")
```
